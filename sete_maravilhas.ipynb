```python
import folium
import branca
```

```python
sete_maravilhas_nomes = ["Grande Pirâmide de Gizé", 
                         "Jardins Suspensos da Babilônia", 
                         "Templo de Ártemis", 
                         "Estátua de Zeus em Olímpia", 
                         "Mausoléu de Halicarnasso", 
                         "Colosso de Rodes",
                         "Farol de Alexandria"
                        ]

sete_maravilhas_coordenadas = [
                                [29.9792, 31.1344],
                                [32.5354, 44.455],
                                [37.9496, 27.3913],
                                [37.6378, 21.6575],
                                [37.0378, 27.4516],
                                [36.451, 28.2553],
                                [31.2138, 29.913]
                              ]

sete_maravilhas_data_construcao = [
                                    "2584–2561 a.C.",
                                    "c. 600 a.C.",
                                    "c. 550 a.C.; e de novo em 323 a.C.",
                                    "466–456 a.C. (templo) 435 a.C. (estátua)",
                                    "351 a.C.",
                                    "292–280 a.C.",
                                    "c. 280 a.C"
                                  ]

sete_maravilhas_imagens = [
                            "https://upload.wikimedia.org/wikipedia/commons/2/2b/Piramida_Cheopsa.jpg",
                            "https://upload.wikimedia.org/wikipedia/commons/a/ae/Hanging_Gardens_of_Babylon.jpg",
                            "https://upload.wikimedia.org/wikipedia/commons/1/1d/Miniaturk_009.jpg",
                            "https://upload.wikimedia.org/wikipedia/commons/6/66/Le_Jupiter_Olympien_ou_l%27art_de_la_sculpture_antique.jpg",
                            "https://upload.wikimedia.org/wikipedia/commons/e/e8/Mausoleum_of_Halicarnassus_by_Ferdinand_Knab_%281886%29.png",
                            "https://upload.wikimedia.org/wikipedia/commons/8/8a/Coloso_de_Rodas..jpg",
                            "https://upload.wikimedia.org/wikipedia/commons/3/33/PHAROS2013-3000x2250.jpg"
                          ]
```

```python
# import tiles from https://leaflet-extras.github.io/leaflet-providers/preview/ 
# icon library https://getbootstrap.com/docs/3.3/components/
```

```python
sete_maravilhas_mapa = folium.Map(location = [35.5322, 25.0928], zoom_start = 5, control_scale=True)
folium.TileLayer(tiles = 'https://basemap.nationalmap.gov/arcgis/rest/services/USGSImageryOnly/MapServer/tile/{z}/{y}/{x}',
                attr = 'Tiles courtesy of the <a href="https://usgs.gov/">U.S. Geological Survey</a>',
                name = 'USGS.USImagery').add_to(sete_maravilhas_mapa)
for i in range(len(sete_maravilhas_coordenadas)):
    coordenada = sete_maravilhas_coordenadas[i]
    maravilha = sete_maravilhas_nomes[i]
    data = sete_maravilhas_data_construcao[i]
    img = sete_maravilhas_imagens[i]
    my_string = "<h4>Nome: {}<h4><h5>Data de Construção: {}<h5><img src='{}' alt='{}' style='height: 200px; width: 300px;'>".format(maravilha, data, img, maravilha)

    iframe = branca.element.IFrame(my_string, width=310, height=350)
    my_popup = folium.Popup(iframe, max_width=360)
    folium.Marker(
            coordenada, 
            popup = my_popup,
            icon = folium.Icon(
                        icon = "glyphicon glyphicon-camera", 
                        color = "black", 
                        icon_color = "white",
                        prefix = "glyphicon")
    ).add_to(sete_maravilhas_mapa)




sete_maravilhas_mapa.add_child(folium.LatLngPopup())

sete_maravilhas_mapa.save("sete_maravilhas.html")

sete_maravilhas_mapa
```
